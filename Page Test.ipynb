```python
def get_Q(traj_num, max_nodes, dtype):
    time_interval, Q = bitlist.read_Q(traj_num, max_nodes, dtype)
    print Q.leaf_number()
    assert Q.leaf_number() == Q.cover_leaf_num(-180, 180, -90, 90)
    return Q

def get_d(traj_num, dtype):
    d = pd.read_pickle('./data/Data_%s_%d.pickle'%(dtype, traj_num))
    return d
```

```python
def cover_page_num(self, x0, x1, y0, y1):
    import math
    if self.is_cover(x0, x1, y0, y1):
        if self.is_leaf():
            return int(math.ceil(1.0 * sum([len(self.nodes[k]) for k in self.nodes.keys()]) / self.page_size))
        else:
            return sum([child.cover_page_num(x0, x1, y0, y1) for child in self.children])
    return 0
quad.QuadTree.cover_page_num = cover_page_num
```

```python
def trajbase(d, Q, points, page_size, width):
    trajs = Q.meet_trajectories_set(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width)
    for p in points[1:]:
        trajs &= Q.meet_trajectories_set(p[0]-width, p[0]+width, p[1]-width, p[1]+width)
    return traj_page_num(d, page_size, trajs)

def traj1(d, Q, points, page_size, width):
    cover_page_num = Q.cover_page_num(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width, page_size)
    ret = Q.cover_trajectory_detail_set(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width)
    pos_traj = ret[0]
    for p in points[1:]:
        ret = Q.cover_trajectory_detail_set(p[0]-width, p[0]+width, p[1]-width, p[1]+width)
        pos_traj &= ret[0]
        if ret[0] in pos_traj:
            cover_page_num += Q.cover_page_num(p[0]-width, p[0]+width, p[1]-width, p[1]+width, page_size)
    return (cover_page_num + traj_page_num(d, page_size, pos_traj))

def traj2(d, Q, points, page_size, width):
    trajs_first = Q.meet_trajectories_set(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width)
    return traj_page_num(d, page_size, trajs_first)

def pos_traj(d, Q, points, page_size, width):
    ret = Q.cover_trajectory_detail_set(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width)
    pos_traj_set = ret[0]
    for p in points[1:]:
        ret = Q.cover_trajectory_detail_set(p[0]-width, p[0]+width, p[1]-width, p[1]+width)
        pos_traj_set &= ret[0]
    return pos_traj_set

def candi_traj(d, Q, points, page_size, width):
    trajs = Q.meet_trajectories_set(points[0][0]-width, points[0][0]+width, points[0][1]-width, points[0][1]+width)
    for p in points[1:]:
        trajs &= Q.meet_trajectories_set(p[0]-width, p[0]+width, p[1]-width, p[1]+width)
    return trajs
```
